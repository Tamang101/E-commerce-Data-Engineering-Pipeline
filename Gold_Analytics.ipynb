Loading  the cleaned Silver Dataset 

```python
orders = pd.read_parquet('C:/Users/asus/Downloads/Project.de/Silver/silver_orders.parquet')
```

```python
customers = pd.read_parquet('C:/Users/asus/Downloads/Project.de/Silver/silver_customers.parquet')
```

```python
Items = pd.read_parquet('C:/Users/asus/Downloads/Project.de/Silver/silver_items.parquett')
```

Join order table with Customers to get order_city

```python
order_city = orders.merge(customers, on='customer_id',how="left")
```

# Revenue per city

```python
Revenue_city = ( order_city
            .groupby('city', as_index = False)['amount']
            .sum()
            .rename(columns={"amount": "Revenue"})
            .sort_values("Revenue",ascending=False)
    
)
```

```python
Revenue_city.head()
```

# Daily orders

```python
orders['Date'] = pd.to_datetime(orders['created_at'], errors='coerce').dt.date
```

```python
Daily_orders = (orders
            .groupby('Date',as_index=False)['order_id']
            .count()
            .rename(columns={'order_id':'Order_Count'})
            .sort_values('Date')
    
)
```

```python
Daily_orders.head()
```

```python
Items.head()
```

# Top products by quantity

```python
Top_product = (Items.groupby('sku', as_index= False)['qty']
                    .sum()
                    .rename(columns={'qty': 'Total_qty'})
                    .sort_values('Total_qty', ascending = False)
)
```
